```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv ("netflix1.csv")
```

```python
sns.countplot(x = 'type',data = df)
plt.title('Count vs Type of Shows')
```

```python
df['country'].value_counts().head(15)
```

```python
plt.figure(figsize = (12,6))
sns.countplot(y = 'country',order = df['country'].value_counts().index[0:15],data = df)
plt.title('Content On Netflix On the Basis Of Origin')
```

```python
plt.figure(figsize = (10,6))
sns.countplot(x = 'release_year' , order = df['release_year'].value_counts().index[0:20],data = df)
plt.title('Yearly Content Production')
```

```python
Rating = df[df['release_year'] == 2021]
top_country = Rating['country'].value_counts().head(10)
plt.figure(figsize = (8,8))
plt.title ('Country wise Content Production In 2021')
plt.pie(top_country, autopct = '%.2f',labels = top_country.index)
```
